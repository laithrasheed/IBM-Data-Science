### TASK 1.4: Use sub plotting to compare the variations in GDP during recession and non-recession period by developing line plots for each period.
<br>Now, you want to find more insights from the data to understand the reason. <br>Plot a two line charts using subplotting to answer:-
#### How did the GDP vary over time during recession and non-recession periods? 
<br>Make use of <u>add_subplot()</u> from Matplotlib for this comparision.


````python
# Set style
sns.set(style="whitegrid")

# Step 1: Separate the data
df_recession = df[df['Recession'] == 1].groupby('Year')['GDP'].mean().reset_index()
df_non_recession = df[df['Recession'] == 0].groupby('Year')['GDP'].mean().reset_index()

# Step 2: Create subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 5), sharey=True)

# Recession plot
sns.lineplot(ax=axes[0], data=df_recession, x='Year', y='GDP', marker='o', color='red')
axes[0].set_title('GDP Trend During Recession')
axes[0].set_xlabel('Year')
axes[0].set_ylabel('Average GDP')
axes[0].grid(True)

# Non-recession plot
sns.lineplot(ax=axes[1], data=df_non_recession, x='Year', y='GDP', marker='o', color='green')
axes[1].set_title('GDP Trend During Non-Recession')
axes[1].set_xlabel('Year')
axes[1].grid(True)

# Adjust layout
plt.suptitle('GDP Trends: Recession vs Non-Recession Periods', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
````

### TASK 1.5: Develop a Bubble plot for displaying the impact of seasonality on Automobile Sales.
<br>How has seasonality impacted the sales, in which months the sales were high or low? Check it for non-recession years to understand the trend

##### Develop a Bubble plot for displaying Automobile Sales for every month and use Seasonality Weight for representing the size of each bubble<br>
Title this plot as 'Seasonality impact on Automobile Sales'


<details><summary>Click here for a hint</summary>

<p>
     You can create Bubble Chart by calling the scatter()
        <br>Pass the 'Month' and 'Automobile_Sales' to the functions as x and y and then use Seasonality weight for size parameter</p>
</details>


````python

# Step 1: Group data by Month and get average Sales & Seasonality
bubble_df = df.groupby('Month')[['Automobile_Sales', 'Seasonality_Weight']].mean().reset_index()

# Step 2: Scale Seasonality_Weight for bubble size (optional tweak)
bubble_df['Bubble_Size'] = bubble_df['Seasonality_Weight'] * 1000  # Adjust factor if bubbles are too small/large

# Step 3: Plot the Bubble Plot
plt.figure(figsize=(10, 6))
plt.scatter(
    x=bubble_df['Month'],
    y=bubble_df['Automobile_Sales'],
    s=bubble_df['Bubble_Size'],  # Bubble size
    alpha=0.6,
    c='skyblue',
    edgecolors='black'
)

# Step 4: Enhance plot
plt.title('Impact of Seasonality on Automobile Sales')
plt.xlabel('Month')
plt.ylabel('Average Automobile Sales')
plt.grid(True)
plt.xticks(range(1, 13))
plt.tight_layout()
plt.show()
````

 ### TASK 1.6: Use the functionality of Matplotlib to develop a scatter plot to identify the correlation between average vehicle price relate to the sales volume during recessions.
 #### From the data, develop a scatter plot to identify if there a correlation between consumer confidence and automobile sales during recession period? 
 <br> Title this plot as 'Consumer Confidence and Automobile Sales during Recessions'

````python
import pandas as pd
import dash
from dash import html, dcc
from dash.dependencies import Input, Output
import plotly.express as px
import datetime as dt

# Create app
app = dash.Dash(__name__)
app.config.suppress_callback_exceptions = True

# Sample read (Replace with your actual data source)
# df = pd.read_csv('your_data.csv')

# Assuming df already exists and is preloaded
# Extract year and month from the Date column
df['Month'] = pd.to_datetime(df['Date']).dt.month_name()
df['Year'] = pd.to_datetime(df['Date']).dt.year

# Layout Section of Dash
app.layout = html.Div(children=[
    html.H1('Automobile Sales Dashboard'),

    # Outer division starts
    html.Div([

        # First inner division: controls
        html.Div([
            html.H2('Select City:'),

            dcc.RadioItems(
                [{"label": city, "value": city} for city in df['City'].unique()],
                value=df['City'].unique()[0],
                id='input-city',
                inline=True
            ),

            html.Div([
                html.H2('Select Year:', style={'margin-top': '20px'}),
                dcc.Dropdown(
                    options=[{'label': str(year), 'value': year} for year in sorted(df['Year'].unique())],
                    value=sorted(df['Year'].unique())[0],
                    id='input-year'
                )
            ]),
        ]),

        # Second inner division for output graphs
        html.Div([
            html.Div([], id='plot1'),
            html.Div([], id='plot2')
        ], style={'display': 'flex', 'justify-content': 'space-between', 'margin-top': '40px'})
    ])
])

# Callback Section
@app.callback(
    [Output(component_id='plot1', component_property='children'),
     Output(component_id='plot2', component_property='children')],
    [Input(component_id='input-city', component_property='value'),
     Input(component_id='input-year', component_property='value')]
)
def reg_year_display(input_city, input_year):
    # Filter data
    region_data = df[df['City'] == input_city]
    y_r_data = region_data[region_data['Year'] == input_year]

    # Plot 1  Average automobile sales by vehicle type (Pie Chart)
    est_data = y_r_data.groupby('Vehicle_Type')['Automobile_Sales'].mean().reset_index()
    fig1 = px.pie(est_data, names='Vehicle_Type', values='Automobile_Sales',
                  title=f"{input_city}: Average Sales by Vehicle Type in {input_year}")

    # Plot 2  Monthly average automobile sales (Bar Chart)
    veg_data = y_r_data.groupby('Month')['Automobile_Sales'].mean().reset_index()
    # Ensure months are ordered properly
    month_order = ['January', 'February', 'March', 'April', 'May', 'June',
                   'July', 'August', 'September', 'October', 'November', 'December']
    veg_data['Month'] = pd.Categorical(veg_data['Month'], categories=month_order, ordered=True)
    veg_data.sort_values('Month', inplace=True)

    fig2 = px.bar(veg_data, x='Month', y='Automobile_Sales',
                  title=f"{input_city}: Monthly Average Automobile Sales in {input_year}")

    return [dcc.Graph(figure=fig1), dcc.Graph(figure=fig2)]

if __name__ == '__main__':
    app.run()
````

````python
import dash
from dash import html, dcc
from dash.dependencies import Input, Output
import pandas as pd
import plotly.express as px

# Sample data simulating the full structure
data = {
    'Year': [2020, 2020, 2021, 2021, 2022, 2022],
    'Vehicle_Type': ['SUV', 'Sedan', 'SUV', 'Truck', 'Sedan', 'Truck'],
    'Recession': [1, 1, 1, 0, 1, 1],
    'Automobile_Sales': [150, 100, 130, 80, 120, 90],
    'Advertising_Expenditure': [2000, 1500, 1800, 1200, 1600, 1400],
}
df = pd.DataFrame(data)

app = dash.Dash(__name__)

year_options = [{'label': str(y), 'value': y} for y in sorted(df['Year'].unique())]

app.layout = html.Div([
    html.H1("Automobile Sales Dashboard"),

    html.Div([
        html.H2("Select Year:"),
        dcc.Dropdown(
            options=year_options,
            value=2021,
            id='input-year'
        )
    ]),

    html.Div(
        id='output-container',
        className='output-area',
        children=[]
    )
])

@app.callback(
    Output('output-container', 'children'),
    Input('input-year', 'value')
)
def update_output(selected_year):
    filtered_df = df[(df['Year'] == selected_year) & (df['Recession'] == 1)]

    # Pie Chart: Average Automobile Sales per Vehicle Type
    pie_fig = px.pie(
        filtered_df,
        names='Vehicle_Type',
        values='Automobile_Sales',
        title=f'Average Automobile Sales by Vehicle Type - {selected_year} (Recession)'
    )

    # Bar Chart: Advertising Expenditure
    avg_ad_exp = df[df['Recession'] == 1].groupby('Year')['Advertising_Expenditure'].mean().reset_index()
    bar_fig = px.bar(
        avg_ad_exp,
        x='Year',
        y='Advertising_Expenditure',
        title='Average Advertising Expenditure During Recession'
    )

    return [
        dcc.Graph(figure=pie_fig),
        dcc.Graph(figure=bar_fig)
    ]

app.run()
````
